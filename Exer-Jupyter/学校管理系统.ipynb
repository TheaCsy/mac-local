```python
Course_list = []

class School(object):
    def __init__(self,schoolname):
        self.schoolname = schoolname
        self.studlist = []
        self.teachlist = []
        
        global Course_list
    
    def hire(self,obj):
        self.teachlist.append(obj.name)
        print("We now hire a NEW teacher {}".format(obj.name))
    
    def enroll(self,obj):
        self.studlist.append(obj.name)
        print("Welcome NEW student {}".format(obj.name))
    

class Grade(School):
    def __init__(self,schoolname,grade_code,grade_course):
        super(Grade,self).__init__(schoolname)  #继承父类的属性
        self.code = grade_code
        self.course = grade_course
        self.members = []
        Course_list.append(self.course)
        
        print("Now we have {0} course of Grade {1} in our {2} school.".format(self.course,self.code,self.schoolname))
    
    def course_info(self):
        print("Course {} is D1/D2/D3".format(self.course))
    
Python = Grade("BJ",1,"Python")
Linux = Grade("CD",3,"Linux")


class SchoolMember(object):
    def __init__(self,name,age,sex,role):
        self.name = name
        self.age = age
        self.sex = sex
        self.role = role
        self.course_list = []
        
        print("Hi. My name is {0}, I am a {1}".format(self.name,self.role))

stu_num = 0
class Students(SchoolMember):
    def __init__(self,name,age,sex,role,course):
        super(Students,self).__init__(name,age,sex,role)
        global stu_num
        stu_num += 1
        stu_id = course.schoolname + "S" + str(course.code) + str(stu_num).zfill(2)  #zfill自动填充至两位数
        
        self.id = stu_id
        self.mark_list = {}
        
    def study(self,course):
        print("I come to study {0}, my student ID is {1}".format(course.course,self.id))
        
    def payment(self,course):
        print("I paid $1000 for {}".format(course.course))
        self.course_list.append(course.course)
        
    def praise(self,obj):
        print("{0} thinks {1} is great.".format(self.name,obj.name))
            
    def markcheck(self):
        for i in self.mark_list.items():
            print(i)
        
    def out(self):
        print("I am leaving.")

tea_num = 0
class Teachers(SchoolMember):
    def __init__(self,name,age,sex,role,course):
        super(Teachers,self).__init__(name,age,sex,role)
        global tea_num
        tea_num += 1
        tea_id = course.schoolname + "T" + str(course.code) + str(tea_num).zfill(2)
        self.id = tea_id
    
    def teach(self,course):
        print("I am here to teach {0}, my ID is {1}".format(course.course,self.id))
    
    def recordmark(self,date,course,obj,level):
        obj.mark_list["Day"+date] = level

a = Students("Abbey",18,"F","student",Python)
Python.enroll(a)
a.study(Python)
a.payment(Python)

b = Students("Ben",22,"M","student",Python)
Python.enroll(b)
b.study(Python)
b.payment(Python)

c = Teachers("Cici",28,"F","teache",Python)
Python.hire(c)
c.teach(Python)
c.recordmark("1",Python,a,"A")
c.recordmark("5",Python,b,"A")

print(b.course_list)
b.markcheck()
b.praise(c)
```
